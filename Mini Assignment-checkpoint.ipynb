```python
import pandas as pd
import numpy as np
import wrds


db = wrds.Connection()


ticker = 'AAPL'


query = f"""
    SELECT 
        gvkey,
        tic,                
        conm,               
        datadate,           
        act,                
        lct,                
        re,                 
        ebit,               
        at,                 
        sale,              
        csho,               
        prcc_f,             
        lt                  
    FROM 
        comp.funda
    WHERE 
        tic = '{ticker}'      
        AND datafmt = 'STD'
        AND consol = 'C'
        AND indfmt = 'INDL'
        AND popsrc = 'D'
        AND datadate >= '2000-01-01'
    ORDER BY 
        datadate
"""


df_raw = db.raw_sql(query, date_cols=['datadate'])


print("data shape:", df_raw.shape)
print(df_raw.head())
```

```python
def calculate_zscore(df):
    df_z = df.copy()
    EPS = 1e-9 
    
   
    df_z['X1'] = (df_z['act'] - df_z['lct']) / (df_z['at'] + EPS)
    df_z['X2'] = df_z['re'] / (df_z['at'] + EPS)
    df_z['X3'] = df_z['ebit'] / (df_z['at'] + EPS)
    df_z['X4'] = (df_z['csho'] * df_z['prcc_f']) / (df_z['lt'] + EPS)
    df_z['X5'] = df_z['sale'] / (df_z['at'] + EPS)
    
    
    df_z['zscore'] = (1.2 * df_z['X1'] +
                      1.4 * df_z['X2'] +
                      3.3 * df_z['X3'] +
                      0.6 * df_z['X4'] +
                      1.0 * df_z['X5'])
    
   
    conditions = [
        (df_z['zscore'] >= 2.99),
        (df_z['zscore'] < 1.81)
    ]
    choices = ['Safe Zone', 'Distress Zone']
    df_z['status'] = np.select(conditions, choices, default='Gray Zone')
    
    return df_z


c = calculate_zscore(df_raw)
print("Data containing Z-scores and statuses：")
print(c[['datadate', 'tic', 'conm', 'zscore', 'status']].tail(10))
```

```python
c.to_excel("companies.xlsx",index=False)
print("Successfuly export companies.xlsx")
```

```python
df_export = c.copy()
df_export['company_name'] = df_export['conm']          
df_export['ticker'] = df_export['tic']                 
df_export['year'] = pd.to_datetime(df_export['datadate']).dt.year  
df_export = df_export[['company_name', 'ticker', 'year', 'zscore', 'status']]

df_export.to_excel(r"C:\Users\shuaichen\Desktop\financial_health_checker\data\companies.xlsx", index=False)

print("Successfully generated the correct fire: companies.xlsx ")
print(df_export.head())
```

```python
import numpy as np

def calculate_zscore(df):
    df_z = df.copy()
    EPS = 1e-9
    
    # Fill NaN with 0 for financial ratios (or drop rows? We'll fill 0)
    numeric_cols = ['act', 'lct', 're', 'ebit', 'at', 'sale', 'csho', 'prcc_f', 'lt']
    df_z[numeric_cols] = df_z[numeric_cols].fillna(0)
    
    # Calculate ratios
    df_z['X1'] = (df_z['act'] - df_z['lct']) / (df_z['at'] + EPS)
    df_z['X2'] = df_z['re'] / (df_z['at'] + EPS)
    df_z['X3'] = df_z['ebit'] / (df_z['at'] + EPS)
    df_z['X4'] = (df_z['csho'] * df_z['prcc_f']) / (df_z['lt'] + EPS)
    df_z['X5'] = df_z['sale'] / (df_z['at'] + EPS)
    
    # Calculate Z-score
    df_z['zscore'] = (1.2 * df_z['X1'] +
                      1.4 * df_z['X2'] +
                      3.3 * df_z['X3'] +
                      0.6 * df_z['X4'] +
                      1.0 * df_z['X5'])
    
    # Create boolean conditions (ensure they are boolean)
    cond_safe = df_z['zscore'] >= 2.99
    cond_distress = df_z['zscore'] < 1.81
    
    # Assign status using numpy where (vectorized)
    df_z['status'] = 'Gray Zone'
    df_z.loc[cond_safe, 'status'] = 'Safe Zone'
    df_z.loc[cond_distress, 'status'] = 'Distress Zone'
    
    return df_z

df_zscore = calculate_zscore(df_raw)

# Add year column
df_zscore['year'] = df_zscore['datadate'].dt.year

# Sector mapping (same as before)
sector_map = {
    'AAPL': 'Tech', 'MSFT': 'Tech', 'GOOGL': 'Tech', 'META': 'Tech', 'NVDA': 'Tech',
    'CSCO': 'Tech', 'INTC': 'Tech', 'ADBE': 'Tech', 'CRM': 'Tech', 'AVGO': 'Tech', 'TXN': 'Tech',
    'AMZN': 'Consumer', 'TSLA': 'Consumer', 'WMT': 'Consumer', 'HD': 'Consumer', 'DIS': 'Consumer',
    'NFLX': 'Consumer', 'PG': 'Consumer', 'PEP': 'Consumer', 'KO': 'Consumer', 'COST': 'Consumer',
    'NKE': 'Consumer', 'T': 'Telco', 'VZ': 'Telco',
    'JPM': 'Financial', 'V': 'Financial',
    'JNJ': 'Health', 'ABT': 'Health', 'TMO': 'Health'
}
df_zscore['sector'] = df_zscore['tic'].map(sector_map).fillna('Other')

# Select columns for export
df_export = df_zscore[['tic', 'conm', 'year', 'zscore', 'status', 'sector']]

# Save to Excel
df_export.to_excel('data/companies_zscore.xlsx', index=False)

print(f"Saved {len(df_export)} rows to data/companies_zscore.xlsx")
print(f"Companies: {df_export['tic'].nunique()}")
print(f"Years: {df_export['year'].min()} - {df_export['year'].max()}")
df_export.head()
```
